# Resize image for training

```python
from PIL import Image
import os
```

Use pillow to resize images to 128x128 for training.
(images larger than 128x128 will be resized to 128x128, images smaller than will be ignored)

```python
# risize an image using pillow
def resize_image(image, size):
    return image.resize(size)

# open an image using pillow
def open_image(image_path):
    return Image.open(image_path)

# save an image using pillow
def save_image(image, image_path):
    image.save(image_path)

# convert an image to grayscale using pillow
def convert_to_grayscale(image):
    return image.convert('L')
```

```python
list_of_images = []
path_to_images = '../../scraping/data'
path_to_processed_images = './processed_images'
# open the all png images in the a folder dynamically
for image in os.listdir(path_to_images):
    if image.endswith('.png'):
        list_of_images.append(open_image(path_to_images +"/" +image))



# print the first image inside the notebook to see if it works
list_of_images[0]
```

```python
list_of_resized_images = []

# resize all the images in the list
for image in list_of_images:
    list_of_resized_images.append(resize_image(image, (128, 128)))


# print the first image inside the notebook to see if it works
list_of_resized_images[0]
```

```python
# convert all the images to grayscale
list_of_grayscale_images = []
for image in list_of_resized_images:
    list_of_grayscale_images.append(convert_to_grayscale(image))

# print the first image inside the notebook to see if it works
list_of_grayscale_images[0]
```

```python
# save all the images pretend the file name is the same
for image in list_of_grayscale_images:
    save_image(image, path_to_processed_images + '/' + image.filename)
```
